```python
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, models
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/Toufic01/diabatics/refs/heads/main/Diabatics/diabete.csv')
```

```python
data.head()
```

```python
df = data.rename(columns={'HeartDiseaseorAttack':'Heart_Attack'})
```

```python
df.drop(['Smoker', 'Heart_Attack','Fruits','Veggies','GenHlth','MentHlth','Stroke'], axis=1,inplace=True)
```

```python
df = df.rename(columns={'HvyAlcoholConsump':'Alcohol_Consump'})
```

```python
df.drop(['CholCheck','DiffWalk'], axis=1,inplace=True)
```

```python
df.isnull().sum()
```

```python
duplicate_rows = df[df.duplicated()]
duplicate_rows
```

```python
new_df = df.drop_duplicates(keep='first')
new_df
```

```python
duplicate_rows = new_df[df.duplicated()]
```

```python
duplicate_rows
```

```python
new_df
```

```python
x = new_df.drop(['Diabetes'],axis=1)
y = new_df[['Diabetes']]
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=.30,random_state=41)
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
model = keras.Sequential(

    [

        keras.layers.Dense(100, activation='relu', input_shape=(8,)),
        keras.layers.Dense(10, activation='sigmoid')


        
    ]
    
)
```

```python
model.compile(
    optimizer = 'adam',
    loss = 'sparse_categorical_crossentropy',
    metrics =['accuracy']
)
```

```python
model.fit(xtrain, ytrain, epochs = 5)
```

```python
test_lost, test_acc = model.evaluate(xtest, ytest)
print(f'Test accuracy: {test_acc}')
print(f'Test lost: {test_lost}')
```

```python
converter = tf.lite.TFLiteConverter.from_keras_model(model)
tflite_model = converter.convert()
```

```python
# Save the model to a .tflite file
with open('diabetes.tflite', 'wb') as f:
    f.write(tflite_model)
```

```python
input_data = np.array([[45, 1, 1, 28.5, 1, 0, 3, 1]])
```

```python
# Predict using the trained model
prediction = model.predict(input_data)

# Output the prediction
print("Prediction probabilities:", prediction)

# If the model is classification-based (like diabetes risk), you might want the class with the highest probability:
predicted_class = np.argmax(prediction, axis=1)  # Returns the class index
print("Predicted class:", predicted_class)
```
